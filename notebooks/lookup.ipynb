```python
import sys; sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
from datasets import ImageSeg
```

```python
dataset = ImageSeg('/tank2/home/public/Dataset_zhe')
print('Dataset length:', len(dataset))
```

```python
fig, axes = plt.subplots(10, 2, figsize=(5, 20))
fig.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.1, hspace=0.1)

for i in range(10):
    img, seg = dataset[i]
    axes[i, 0].imshow(img)
    axes[i, 1].imshow(seg)
    axes[i, 0].axis('off')
    axes[i, 1].axis('off')

plt.show()
```
